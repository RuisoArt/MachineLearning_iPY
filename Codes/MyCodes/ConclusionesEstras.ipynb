# <span style="color:red">Linear Regression</span>

La regresion lineal como su nombre lo dicta puede ser utilizada para los datos ordenados naturalmente alrrededor de una recta sin estar muy dispersados que digamos.

Es un algoritmos que necesita etiquetas (aprendizaje supervizado).

Sigue la formula clasica de Y = ax + b

Con este algoritmo se pueden realizar prediciones y estimaciones que  cumplan con su representacion.

Puede ubicarse la pendiente de un modelo y la intereseccion como parametros guardados dentro de SK-Learn Linear Regression.

Este modelo **se puede ampliar** para regresiones **NO LINEALES** para esto se transforman los datos con funciones basicas, para que asi se de una regresion no lineal.(Polinomial Regression).

Un patron muy util al utilizar **NO LINEALES** como funcion para hacer un Polinomial Regression es el de una suma Gausiana.

En lo anterior hay que tener **SUMO CUIDADO** pues la introducción de funciones básicas en nuestra regresión lineal hace que el modelo sea mucho más flexible, pero también puede conducir rápidamente a un ajuste excesivo.

Con este algoritmo tambien se puede hacer **Regularizacion de Cresta** o tambien llamada regularización de Tikhonov.

Con este algoritmo tambien se puede hacer **Regularizacion de Lazo**, que, aunque conceptualmente son muy similares a la regresión de cresta, los resultados pueden diferir debido a razones geométricas, la regresión de lazo tiende a favorecer los modelos dispersos cuando es posible, es decir, preferentemente establece los coeficientes del modelo exactamente en cero.

# <span style="color:red">Principal Component Analysis</span>

Ojala hubiese conocido este algoritmo en la adquisicion de ondas del cuerpo humano mediante tarjetas electronicas, pues este permite apoyar a los circuitos de filtrado y suavizado de ondas, con suavizado mediante software sobre ruido adquirido, estimando de mejor manera picos de ondas especificas y de importancia produsemental.

Este es vez de ser un algoritmo supervizado, que utiliza etiquetas para su aprendizaje, estimacion y clasificacion; este es un algoritmo de caracter **NO SUPERVIZADO**.

PCA es fundamentalmente un algoritmo de reducción de dimensionalidad, pero también puede ser útil como herramienta para visualización, filtrado de ruido, extracción de características e ingeniería, y mucho más. 

Puede ser usado en su forma mas basica para ecuaciones lineales reemplazando a el Linear Regression. En lugar de intentar predecir los valores de y a partir de los valores de x, el problema de aprendizaje no supervisado intenta aprender sobre la relación entre los valores de x e y. En el análisis de componentes principales, esta relación se cuantifica encontrando una lista de los ejes principales en los datos y usando esos ejes para describir el conjunto de datos. 

Un uso muy comun para PCA es la reuccion de dimensiones de los datos ( Este conjunto de datos de dimensión reducida es, en algunos sentidos, "lo suficientemente bueno" para codificar las relaciones más importantes entre los puntos: a pesar de reducir la dimensión de los datos en un 50 %, la relación general entre los puntos de datos se conserva en su mayor parte).

# <span style="color:red">k-Means Clustering</span>

Este tipo de algoritmo no supervizado trabaja con agrupamientos en ves de reduccion de dimencionalidades.

Con este algoritmo se pueden rescaar atributos de imagenes por color, saturacion, luces y sombras, etc. Dependiendo lo que se necesite, pues los pixeles de las imagenes se agruparan por color o semejanza de valor.

Este algoritmo se enacarga de buscar un numero predeterminado de grupos dentro de un conjunto de datos multidimencional sin etiquetar; esto lo logra utilizando una concepcion simple de como se ve el agrupamient optimo.

En situaciones donde se repitan valores en matrices numericas, se puede reducir o comprimir el valor para optimizar el agrupamiento de datos, como en la compresion de imagenes.

Una de sus variaciones mas utilizadas en la ciencia de datos es el K-means Expectrarion Maximization mediante el uso de Pasos E que asigna puntos al centro de los clusteres encontrados mas cercanos y el M que toma la media de estos clusteres. Sin emabrgo tiene varios problemas como:

1. Aunque se garantiza que el procedimiento E-M mejorará el resultado en cada paso, no hay garantía de que conduzca a la mejor solución global.
2. Debe decirle cuántos grupos espera: no puede aprender el número de grupos de los datos.
3. k-means se limita a los límites del clúster lineal. Los límites entre los grupos de k-means siempre serán lineales, lo que significa que fallará para límites más complicados.

Para solucionar por lo menos el ultimo problema tenemos que nos debemos dirgir a un uso convinado con SVM para proyectar los datos a una dimensión superior donde es posible una separación lineal.

4. k-means puede ser lento para un gran número de muestras debido a que cada iteración de k -means debe acceder a todos los puntos del conjunto de datos, el algoritmo puede ser relativamente lento a medida que aumenta la cantidad de muestras.
