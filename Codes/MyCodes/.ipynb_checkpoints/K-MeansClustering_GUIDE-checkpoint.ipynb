El algoritmo de $K-Means$ se enacarga de buscar un numero predeterminado de grupos dentro de un conjunto de datos multidimencional sin etiquetar; esto lo logra utilizando una concepcion simple de como se ve el agrupamient optimo.

1. El centro de grupo o **cluster center** es la medida aritmatica de todos los puntos que pertencen al grupo.
2. Cada punto esta mas cerca de su propio centro de conglomerado que de otros centros de conglomerados.

Esos dos supuestos son la base del modelo de $K-Means$.

Primero, generemos un conjunto de datos bidimensional que contenga cuatro blobs distintos. Para enfatizar que este es un algoritmo no supervisado, dejaremos las etiquetas fuera de la visualizacin.

A simple vista, es relativamente fcil distinguir los cuatro grupos. El algoritmo k -means hace esto automticamente, y en Scikit-Learn usa la tpica API de estimador:

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Visualicemos los resultados trazando los datos coloreados por estas etiquetas. Tambin representaremos grficamente los centros de conglomerados determinados por el estimador de $k-Means$:

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

La buena noticia es que el algoritmo k -means (al menos en este caso simple) asigna los puntos a los grupos de manera muy similar a cmo los asignaramos a simple vista.

Afortunadamente para nosotros, una bsqueda tan exhaustiva no es necesaria: en cambio, el enfoque tpico de $K-means$ implica un enfoque iterativo intuitivo conocido como expectativa-maximizacin (**expectation-maximization**)

# K-Means: Expectarion-Maximization

Esta clase de algoritmo, utilizado ampliamente en la ciencia de datos, consiste el siguiente procedimiento como algoritmo:

1. Adivina algunos centros de racimo
2. Repetir hasta converger
    a. E-Step : asigna puntos al centro del clster ms cercano
    b. M-Step : establezca los centros de los clsteres en la media
    
Aqu, el "**Step E**" o "paso de expectativa" se llama as porque implica actualizar nuestra expectativa de a qu grupo pertenece cada punto.

El "**Step M**" o "paso de maximizacin" se llama as porque implica maximizar alguna funcin de aptitud que define la ubicacin de los centros del conglomerado; en este caso, esa maximizacin se logra tomando una media simple de los datos en cada conglomerado.

En circunstancias tpicas, cada repeticin del **Step E** y del **Step M** siempre dar como resultado una mejor estimacin de las caractersticas del grupo.

Podemos visualizar el algoritmo como se muestra en la siguiente figura. Para la inicializacin particular que se muestra aqu, los clsteres convergen en solo tres iteraciones.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()  # for plot styling
import numpy as np

from ipywidgets import interact
from sklearn.metrics import pairwise_distances_argmin
from sklearn.datasets import make_blobs

def plot_kmeans_interactive(min_clusters=1, max_clusters=6):
    X, y = make_blobs(n_samples=300, centers=4,
                      random_state=0, cluster_std=0.60)
        
    def plot_points(X, labels, n_clusters):
        plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis',
                    vmin=0, vmax=n_clusters - 1);
            
    def plot_centers(centers):
        plt.scatter(centers[:, 0], centers[:, 1], marker='o',
                    c=np.arange(centers.shape[0]),
                    s=200, cmap='viridis')
        plt.scatter(centers[:, 0], centers[:, 1], marker='o',
                    c='black', s=50)
            

    def _kmeans_step(frame=0, n_clusters=4):
        rng = np.random.RandomState(2)
        labels = np.zeros(X.shape[0])
        centers = rng.randn(n_clusters, 2)

        nsteps = frame // 3

        for i in range(nsteps + 1):
            old_centers = centers
            if i < nsteps or frame % 3 > 0:
                labels = pairwise_distances_argmin(X, centers)

            if i < nsteps or frame % 3 > 1:
                centers = np.array([X[labels == j].mean(0)
                                    for j in range(n_clusters)])
                nans = np.isnan(centers)
                centers[nans] = old_centers[nans]

        # plot the data and cluster centers
        plot_points(X, labels, n_clusters)
        plot_centers(old_centers)

        # plot new centers if third frame
        if frame % 3 == 2:
            for i in range(n_clusters):
                plt.annotate('', centers[i], old_centers[i], 
                             arrowprops=dict(arrowstyle='->', linewidth=1))
            plot_centers(centers)

        plt.xlim(-4, 4)
        plt.ylim(-2, 10)

        if frame % 3 == 1:
            plt.text(3.8, 9.5, "1. Reassign points to nearest centroid",
                     ha='right', va='top', size=14)
        elif frame % 3 == 2:
            plt.text(3.8, 9.5, "2. Update centroids to cluster means",
                     ha='right', va='top', size=14)
    
    return interact(_kmeans_step, frame=[0, 50],
                    n_clusters=[min_clusters, max_clusters])

plot_kmeans_interactive();
```

El algoritmo k -Means es tan simple que podemos escribirlo en unas pocas lneas de cdigo. La siguiente es una implementacin muy bsica:

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis')
```

La mayora de las implementaciones bien probadas harn un poco ms que esto bajo el cap, pero la funcin anterior proporciona la esencia del enfoque de maximizacin de expectativas.

# K-Means: Caveats of expectationmaximization

Hay algunos problemas a tener en cuenta al usar el algoritmo de maximizacin de expectativas.

### **Es posible que no se logre el resultado globalmente ptimo**. 

Primero, aunque se garantiza que el procedimiento E-M mejorar el resultado en cada paso, no hay garanta de que conduzca a la mejor solucin global.

Por ejemplo, si usamos una semilla aleatoria diferente en nuestro procedimiento simple, las conjeturas iniciales particulares conducen a resultados deficientes:

```python
centers, labels = find_clusters(X, 4, rseed=0)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis')
```

Aqu el enfoque E-M ha convergido, pero no ha convergido a una configuracin globalmente ptima. Por esta razn, es comn que el algoritmo se ejecute para varias conjeturas iniciales, como lo hace Scikit-Learn de manera predeterminada (establecido por el $n - init$ parmetro, que tiene un valor predeterminado de 10).

### **El nmero de clsteres debe seleccionarse previamente**

Otro desafo comn con k -means es que debe decirle cuntos grupos espera: no puede aprender el nmero de grupos de los datos. 

Por ejemplo, si le pedimos al algoritmo que identifique seis grupos, proceder felizmente y encontrar los mejores seis grupos.

```python
labels = KMeans(6, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

Si el resultado es significativo es una pregunta difcil de responder de manera definitiva; un enfoque que es bastante intuitivo es utilizar ** silhouette analysis** pero no es tema de K-means como para verlo aqui, sin embargo lo puede ver en el siguiente enlace:

https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

Alternativamente, puede usar un algoritmo de agrupamiento ms complicado que tenga una mejor medida cuantitativa de la aptitud por nmero de conglomerados o que pueda elegir una cantidad adecuada de conglomerados.

### k-means se limita a los lmites del clster lineal.

Los supuestos fundamentales del modelo de k -means (los puntos estarn ms cerca de su propio centro de conglomerado que de otros) significa que el algoritmo a menudo ser ineficaz si los conglomerados tienen geometras complicadas.

En particular, los lmites entre los grupos de k-means siempre sern lineales, lo que significa que fallar para lmites ms complicados. Considere los siguientes datos, junto con las etiquetas de conglomerados encontradas por el enfoque tpico de k-means:

```python
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(200, noise=.05, random_state=0)
```

```python
labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis');
```

En este tipo de casos, remitiendonos a **SVM** usamos una transformacin del ncleo para proyectar los datos a una dimensin superior donde es posible una separacin lineal.

Podramos imaginarnos usando el mismo truco para permitir que k -means descubra lmites no lineales.

Una versin de este k-means kernelizado se implementa en Scikit-Learn dentro del $SpectralClusteringestimador$. Utiliza el grfico de los vecinos ms cercanos para calcular una representacin de mayor dimensin de los datos y luego asigna etiquetas usando un algoritmo k -means:

```python
from sklearn.cluster import SpectralClustering
```

```python
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis');
```

Vemos que con este enfoque de transformacin del kernel, el k -means kernelizado puede encontrar los lmites no lineales ms complicados entre los clsteres.

### k-means puede ser lento para un gran nmero de muestras 

Debido a que cada iteracin de k -means debe acceder a todos los puntos del conjunto de datos, el algoritmo puede ser relativamente lento a medida que aumenta la cantidad de muestras.

Quizs se pregunte si este requisito de usar todos los datos en cada iteracin se puede relajar.

Por ejemplo, puede usar un subconjunto de los datos para actualizar los centros de clster en cada paso. Esta es la idea detrs de los algoritmos k - means basados en lotes, una forma de la cual se implementa en $sklearn.cluster.MiniBatchKMeans$.
